When you use df.iloc[[0]], you are accessing the first row of the DataFrame df, but the result is a DataFrame instead of a Series.

# Access specific rows and columns of a pandas DataFrame with .iloc indexer

## df

This is a pandas DataFrame object.

## .iloc

This is a property of DataFrame used for integer-location based indexing.

    [[0, 4], [0, 1, 2]]:

The first list [0, 4] specifies the row indices.

The second list [0, 1, 2] specifies the column indices.

Result of df.iloc[[0, 4], [0, 1, 2]]

This will select the intersection of the specified rows and columns. It means that:

Rows at positions 0 and 4.

Columns at positions 0, 1, and 2.

```python
df.iloc[:,[0,1,2]].head()   # Here : means all rows
```

```python
df5 = df.iloc[0:5,0:3]  # First two rows and first three columns
df5
```

# Complex Dataframe Slicing
## df

This is a pandas DataFrame object. It contains the data from which we want to select a subset.

## .iloc

This is a property of the DataFrame used for integer-location based indexing. It allows you to select rows and columns by their integer positions.

##0:5

This specifies the row slice.

0:5 means rows from index 0 up to, but not including, index 5.

This will select the first 5 rows (indices 0, 1, 2, 3, and 4).

##0:3

This specifies the column slice.

0:3 means columns from index 0 up to, but not including, index 3.

This will select the first 3 columns (indices 0, 1, and 2).

Result of df.iloc[0:5, 0:3]

The result is a new DataFrame df5 that contains the first 5 rows and the first 3 columns of the original DataFrame df.

```python
df5.describe()
```
